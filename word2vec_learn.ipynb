```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
```

```python
import os
```

```python
from gensim.parsing.preprocessing import remove_stopwords
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords
import nltk 
nltk.download('punkt')
```

```python
story = []
filename = '/content/drive/MyDrive/1_10_seasons_tbbt.csv'
f = open(os.path.join('data',filename))
corpus = f.read()
raw_sent = sent_tokenize(corpus)
for sent in raw_sent:
  story.append(simple_preprocess(sent))
  story.append(remove_stopwords(sent))
```

```python
len(story)
```

```python
model = gensim.models.Word2Vec(window=10, min_count=2)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count,epochs=20)
```

```python
model.wv.most_similar('howard')
```

```python
model.wv.most_similar('emily')
```

```python
model.wv.doesnt_match(['raj', 'amy', 'penny', 'bernadette'])
```
